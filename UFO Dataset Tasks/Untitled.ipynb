```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv("scrubbed.csv")
```

```python
df
```

```python
df = df[df['country'] == 'us']
```

```python
df
```

```python
df["datetime"] = pd.to_datetime(df["datetime"], format="mixed", errors="coerce")
df
```

```python
df = df.dropna()
df
```

```python
df2000 = df[df.datetime.dt.year>1990]
df2000 = df2000.reset_index(drop=True)
df2000.loc[:, 'year'] = df2000.datetime.dt.year.astype(str)
df_final = df2000.groupby("year").apply(lambda x: x.sample(frac=0.08)).reset_index(drop=True)
df_final
```

```python
df_final.groupby("year").count()
```
